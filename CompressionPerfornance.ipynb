```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def calculate_psnr_ssim(video1_path, video2_path):
    # Open video files
    cap1 = cv2.VideoCapture(video1_path)
    cap2 = cv2.VideoCapture(video2_path)

    # Get video properties
    frame_count1 = int(cap1.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_count2 = int(cap2.get(cv2.CAP_PROP_FRAME_COUNT))

    # Ensure both videos have the same frame count
    assert frame_count1 == frame_count2, "Videos must have the same frame count"

    total_psnr = 0
    total_ssim = 0
    frame_idx = 0

    while True:
        # Read frames from both videos
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()

        if not ret1 or not ret2:  # Exit loop if either video ends
            break

        # Convert frames to grayscale for SSIM calculation
        gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
        gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

        # Calculate PSNR
        psnr_value = cv2.PSNR(frame1, frame2)
        total_psnr += psnr_value

        # Calculate SSIM
        ssim_value, _ = ssim(gray1, gray2, full=True)
        total_ssim += ssim_value

        frame_idx += 1

    # Release video resources
    cap1.release()
    cap2.release()

    # Calculate averages
    avg_psnr = total_psnr / frame_idx if frame_idx > 0 else 0
    avg_ssim = total_ssim / frame_idx if frame_idx > 0 else 0

    print(f"Average PSNR: {avg_psnr:.2f} dB")
    print(f"Average SSIM: {avg_ssim:.4f}")

    return avg_psnr, avg_ssim

# # Example usage
# video1 = "result/pexels-koolshooters-7322716/blend_0p0025.mp4"
# video2 = "result/pexels-koolshooters-7322716/blend.mp4"
# average_psnr, average_ssim = calculate_psnr_ssim(video1, video2)

# print(f"Results:\nPSNR: {average_psnr:.2f} dB\nSSIM: {average_ssim:.4f}")
```
